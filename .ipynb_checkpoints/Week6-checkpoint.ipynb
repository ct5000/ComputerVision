```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Exercise 6.1
def gaussian1DKernel(sigma):
    filter_size = np.ceil(5*sigma)
    x_row = np.arange(-filter_size,(filter_size)+1)
    #g_1 = g_1[:,np.newaxis]
    g = 1/(np.sqrt(2*np.pi*sigma**2))* np.exp((-x_row**2)/(2*sigma**2))
    gx = -(-x_row**2/(sigma**2)) * g
    g = g[:,np.newaxis]
    gx = gx[:,np.newaxis]
    return g, gx
    
    
    
g, gx = gaussian1DKernel(3)
print(sum(g))
print(gx)
```

```python
# Exercise 6.2
def gaussianSmoothing(im,sigma):
    g,gx = gaussian1DKernel(sigma)
    I = cv2.filter2D(cv2.filter2D(im, -1, g), -1, g.T) 
    Ix = cv2.filter2D(cv2.filter2D(im, -1, gx.T), -1, g)
    Iy = cv2.filter2D(cv2.filter2D(im, -1, g.T), -1, gx)
    
    #I = (g@g.T)@im
    #Ix = (g@gx.T)@im
    #Iy = (gx@g.T)@im
    
    return I, Ix, Iy

img = cv2.imread('week06_data/TestIm1.png')
#img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
I,Ix,Iy = gaussianSmoothing(img,10)
plt.imshow(I)
plt.show()

```

```python
# Exercise 6.3

def smoothedHessian(im, sigma, epsilon):
    I, Ix, Iy = gaussianSmoothing(im,sigma)
    g_eps, gx = gaussian1DKernel(epsilon)
    C = np.array([[cv2.filter2D(cv2.filter2D(Ix**2,-1,g_eps),-1,g_eps.T),cv2.filter2D(cv2.filter2D(Ix*Iy,-1,g_eps),-1,g_eps.T)],
                  [cv2.filter2D(cv2.filter2D(Ix*Iy,-1,g_eps),-1,g_eps.T),cv2.filter2D(cv2.filter2D(Iy**2,-1,g_eps),-1,g_eps.T)]])
    
    
    
    return C
C = smoothedHessian(img,10,3)
                  

print(C.shape)
```

```python
# Exercise 6.4

def harrisMeasure(im,sigma,epsilon,k):
    C = smoothedHessian(im,sigma,epsilon)
    a, b, c = C[0, 0, :, :], C[1, 1, :, :], C[0, 1, :, :]
    r = a*b - c**2 - k * (a + b)**2
    return r


sigma=0.7
r = harrisMeasure(img, sigma=sigma, epsilon=np.ceil(sigma), k=0.06)
plt.imshow(r)
print(r.shape)
```

```python
# Exercise 6.5
def cornerDetector(im,sigma,epsilon,k,tau):
    r = harrisMeasure(im,sigma,epsilon,k)
    c = []
    for i in range(1,r.shape[0]-1):
        for j in range(1,r.shape[1]-1):
            val = r[i,j]
            if val > tau:
                if r[i+1,j] > val and r[i-1,j] > val and r[i,j+1] > val and r[i,j-1] > val:
                    c.append([i,j])
    return c

c = cornerDetector(img, sigma, np.ceil(sigma), 0.06, 100)
print(c)

```
